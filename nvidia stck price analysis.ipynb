```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\harsh\OneDrive\Desktop\projects\NVIDIA_STOCK.csv")
```

```python
print(df['Close'].dtype)
```

```python
print(df['Close'].head())
```

```python
df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
```

```python
df = df.dropna(subset=['Close'])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Close'], label='Close Price', color='blue')
plt.title('Close Price Trend')
plt.xlabel('Index')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
```

```python
correlation_matrix = numeric_df.corr()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.histplot(df['Volume'], kde=True, color='green')
plt.title('Volume Distribution')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.show()
```

```python
highest_close = df['Close'].max()
lowest_close = df['Close'].min()

print(f"Highest Close Price: {highest_close}")
print(f"Lowest Close Price: {lowest_close}")
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['High'], label='High', color='green')
plt.plot(df['Low'], label='Low', color='red')
plt.plot(df['Open'], label='Open', color='purple')
plt.title('High, Low, and Open Prices')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
sns.scatterplot(x='Volume', y='Price', data=df, color='orange')
plt.title('Volume vs Price')
plt.xlabel('Volume')
plt.ylabel('Price')
plt.show()
```
